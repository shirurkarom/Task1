```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
# Load your dataset into a Pandas DataFrame
df =d1
# Create a LabelEncoder object
le = LabelEncoder()
# Apply the LabelEncoder to your categorical column
df['Artist'] = le.fit_transform(df['Artist'])
df['Album'] = le.fit_transform(df['Album'])
df['Genre'] = le.fit_transform(df['Genre'])
df['Release Date'] = le.fit_transform(df['Release Date'])
df['Duration'] = df['Duration'].round().astype(int)
df['Popularity'] = df['Popularity'].round().astype(int)
print(df)
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        sns.countplot(x=column, data=df)
        plt.title(f'Distribution of {column}')
        plt.show()
    else:
        sns.distplot(df[column], kde=True)
        plt.title(f'Distribution of {column}')
        plt.show()
```

```python
# Check for outliers
for column in df.columns:
    if df[column].dtype != 'object':
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        print(f'Outliers in {column}: {outliers.shape[0]}')
```

```python
# Plot the outliers
for column in df.columns:
    if df[column].dtype != 'object':
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        sns.boxplot(x=df[column])
        plt.title(f'Outliers in {column}')
        plt.show()
```

```python
from scipy import stats
```

```python
# Perform statistical tests to check for normality
for column in df.columns:
    if df[column].dtype != 'object':
        stat, p = stats.normaltest(df[column])
        print(f'Normality test for {column}: p-value = {p}')
        if p < 0.05:
            print(f'{column} is not normally distributed')
        else:
            print(f'{column} is normally distributed')
```
